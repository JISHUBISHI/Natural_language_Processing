```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam (1).csv")
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df["Category"].apply(lambda x:1 if x =='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam,test_size=0.20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
type(X_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_CV = v.fit_transform(X_train.values)
X_train_CV
```

```python
X_train_CV.toarray()[:2]
```

```python
X_train_CV.shape
```

```python
v.get_feature_names_out()[6847]
```

```python
v.vocabulary_
```

```python
X_train_np = X_train_CV.toarray()
X_train_np[:4]
np.where(X_train_np[0]!= 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train_CV,y_train)
```

```python
X_test_CV = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_CV)
classification_report(y_test,y_pred)
```

```python
emails = ['Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat...'
          'Yeah do! Don‘t stand to close tho- you‘ll catch something!']
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```
